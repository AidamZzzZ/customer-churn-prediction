```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/home/usuario/Escritorio/customer-churn-prediction/data/raw/train.csv'
churn_data = pd.read_csv(path)
```

## Tipos de datos
- id: Identificacion
- CustomerID: Itentificacion del cliente 
- Surname: Apellid del cliente
- CreditScore: Puntaje de credito del cliente (Numerico:continuo)
- Geography: Pais donde se encuentra el cliente (Categorico:nominal)
- Gender: Genero del cliente (Categorico:nominal)
- Age: Edad del cliente (Numerico:discreto)
- Tenure: Numero de anos que el cliente lleva el el banco (Numerico:discreto)
- Balance: Saldo actual en la cuenta del cliente (Numerico:continuo)
- NumOfProducts: Numero de productos que el cliente tiene con el banco (Numerico:discreto)
- HasCrCar: El cliente tiene o no cuenta de credito con el banco (Categorico:binario)
- IsActiveMember: Es miembro activo del banco (Categorico:Binario)
- EstimatedSalary: Salario estimado (Numerico:continuo)
- Exited: El cliente sigue en siendo cliente del banco (Categorico:binario)

## Analisis de datos

```python
# Distribucion de credito
sns.histplot(churn_data['CreditScore'])
plt.title("Distribucion de credito")
plt.show()
```

Gran parte de los clientes suelen tener una buena puntuacion de credito de alrededor  de 600 a 750 puntos aproximada mente

```python
# distribucion de balance en la cuenta de cada cliente
sns.histplot(churn_data['Balance'])
plt.title('Distribucion de balance en cuentas bancarias')
plt.show()
```

En su mayoria los clientes tienen 0 dolares en su cuenta, y una pequena parte tiene enter $100k a $150k

en su cuenta bancaria

```python
# Distribucion anual de salario de clientes
sns.histplot(churn_data['EstimatedSalary'])
plt.title('Distribucion estimada de salario')
plt.show()
```

Por lo general los clientes suelen tenerrr salarios de entre $50k a $180k aproximadamente

```python
sns.countplot(x='Tenure', data=churn_data)
plt.title('Tiempo de cada cliente en el banco')
plt.ylabel('Frecuencia')
plt.show()
```

Se ve una distribucion unifome entre las personas que tienen 1 a 9 años

```python
# Distribucion de edades
plt.figure(figsize=(15, 8))
sns.countplot(data=churn_data, x='Age')
plt.title('Distribucion de edades')
plt.xticks(rotation=45)
plt.show()
```

La mayoria de clientes tiene entre 30 a 45 años de edad

```python
# Graficacion de productos financieros
sns.countplot(data=churn_data, x='NumOfProducts')
plt.title("Productos financieros por cliente")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(8, 5))

sns.countplot(x='HasCrCard', data=churn_data, ax=axes[0])
axes[0].set_title('Tiene tarjeta de credito')

sns.countplot(x='IsActiveMember', data=churn_data, ax=axes[1])
axes[1].set_title('Es un cliente activo')

sns.countplot(x='Exited', data=churn_data, ax=axes[2])
axes[2].set_title('Sigue siendo cliente del banco')
plt.tight_layout()
plt.show()
```
