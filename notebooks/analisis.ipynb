## Tipos de datos
- id: Identificacion
- CustomerID: Itentificacion del cliente 
- Surname: Apellid del cliente
- CreditScore: Puntaje de credito del cliente (Numerico:continuo)
- Geography: Pais donde se encuentra el cliente (Categorico:nominal)
- Gender: Genero del cliente (Categorico:nominal)
- Age: Edad del cliente (Numerico:discreto)
- Tenure: Numero de anos que el cliente lleva el el banco (Numerico:discreto)
- Balance: Saldo actual en la cuenta del cliente (Numerico:continuo)
- NumOfProducts: Numero de productos que el cliente tiene con el banco (Numerico:discreto)
- HasCrCar: El cliente tiene o no cuenta de credito con el banco (Categorico:binario)
- IsActiveMember: Es miembro activo del banco (Categorico:Binario)
- EstimatedSalary: Salario estimado (Numerico:continuo)
- Exited: El cliente sigue en siendo cliente del banco (Categorico:binario)
